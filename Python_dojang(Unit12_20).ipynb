## **Unit 12. 딕셔너리 사용하기**

- 연관된 값을 묶어서 저장하는 용도로 사용하는 자료형
- 해시 기법을 이용해서 데이터 저장
- 딕셔너리와 같은 키-값 형태의 자료형 == 해시, 해시 맵, 해시 테이블

        : {}, dict() (빈 딕셔너리)
        : 딕셔너리 = {키1:값1, 키2:값2}
        : dict(키1=값1, 키2=값2)
        : dict(zip([키1, 키2], [값1, 값2]))
        : dict([(키1, 값1), (키2, 값2)])
        : dict({키1: 값1, 키2: 값2})
        : 키에는 값을 하나만 지정 가능
        : 키는 중복 불가

        : 딕셔너리의 값 모든 자료형 사용 가능

## **Unit 18. break, continue로 반복문 제거하기**

- break는 for와 while 문법에서 제어흐름을 벗어나기 위해 사용
- continue는 제어흐름을 유지한 상태에서 코드의 실행만 건너뛰는 역할

18.1 break로 반복문 끝내기

```python
i = 0
while True : 
  print(i)
  i += 1
  if i == 10:
      break # 반복문을 끝냄, while의 제어흐름을 벗어남
```

18.2 continue로 코드 실행 건너뛰기

```python
for i in range(10):
  if i % 2 ==0 :
    continue # 아래 코드를 실행하지 않고 건너뜀
  print(i)


i = 0
while i < 15:
  i += 1
  if i % 2 == 0 :
    continue
  print(i) 
```

## **Unit 19. 계단식으로 별 출력하기**

19.1 중첩 루프 
- 반복문 안에 반복문
- 변수 : i, j, k
- 2차원 평면을 다룰 수 있으며, 이미지 처리, 영상처리, 좌표계 처리등에 사용
      for i in range(횟수): # 바깥쪽 루프
        for j in range(횟수): # 안쪽 루프
          처리코드 (i)

```python
for i in range(5):
  for j in range(5):
    print('j:',j, sep='', end=' ')
  print('i: ',i, '\\n', sep='')
```

```python
for i in range(5):
  for j in range(5):
    print('*', end='')
  print()
```

```python
for i in range(5):
  for j in range(5):
    if j <= i:
      print('*', end='')
  print()
```

```python
for i in range(5):
  for j in range(5):
    if j == i :
      print('*', end ='')
    else:
      print(' ', end='')
  print()
```

```python
h = int(input())

for i in range(h):
  for j in reversed(range(h)):
    if j > i *:
      print(' ', end='')
    else:
      print('*', end='')
  
  for j in range(h):
    if j < i :
      print('*', end='')
  print()
```

20.3 3과 5의 공배수 처리

```python
for i in range(1,101):
  if i % 3 == 0 and i % 5 == 0 : # 3과 5의 공배수를 먼저 검사 한 뒤, 각 각의 배수에 대해 검사해야함
    print('FizzBuzz') # 그렇지 않을 경우, 공배수 처리를 하지 않고 3의 배수로 나오기 때문
  elif i % 3 == 0 :
    print('Fizz')
  elif i % 5 == 0 :
    print('Buzz')
  else:
    print(i)
```

20.4 논리 연산자를 사용하지 않고 3과 5의 공배수 처리

```python
for i in range(1,101):
  if i % 15 == 0: # 주석이 없다면 최소공배수임을 알기 힘듦. 가독성이 높은 위의 식 추천
    print('FizzBuzz')
  elif i % 3 == 0:
    print('Fizz')
  elif i % 5 == 0:
    print('Buzz')
  else: 
    print(i)
```

연습문제) <br>
표준 입력으로 정수 두 개가 입력됩니다(첫 번째 입력 값의 범위는 1-1000, 두 번째 입력 값의 범위는 10-1000이며 첫 번째 입력 값은 두 번째 입력 값보다 항상 작습니다). 첫 번째 정수부터 두 번째 정수까지 숫자를 출력하면서 5의 배수일 때는 'Fizz', 7의 배수일 때는 'Buzz', 5와 7의 공배수일 때는 'FizzBuzz'를 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다).

```python
a, b = map(int,input().split())

for i in range(a,b+1):
  if i % 5 ==0 and i % 7 == 0 :
    print('FizzBuzz')
  elif i % 5 == 0 :
    print('Fizz')
  elif i % 7 == 0 :
    print('Buzz')
  else:
    print(i)
```
